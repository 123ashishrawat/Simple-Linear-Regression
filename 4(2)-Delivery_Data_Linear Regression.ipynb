```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Load the dataset
data = pd.read_csv("delivery_time.csv")
df = pd.DataFrame(data)
```

```python
# EDA - Explore the data
plt.scatter(df['Sorting Time'], df['Delivery Time'])
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.title('Sorting Time vs. Delivery Time')
plt.show()
```

```python
# Create the feature matrix (X) and target variable (y)
X_delivery = df[['Sorting Time']]
y_delivery = df['Delivery Time']
```

```python
# Build the linear regression model
model_delivery = LinearRegression()
model_delivery.fit(X_delivery, y_delivery)
```

```python
# Make predictions
predictions_delivery = model_delivery.predict(X_delivery)
```

```python
# Calculate the Mean Squared Error (MSE) for the original model
mse_delivery = mean_squared_error(y_delivery, predictions_delivery)
print(f"Mean Squared Error for Delivery_time (Original Data): {mse_delivery}")
```

```python
# Log transformation
X_log = np.log(X_delivery)
model_delivery.fit(X_log, y_delivery)
predictions_log = model_delivery.predict(X_log)
rmse_log = np.sqrt(mean_squared_error(y_delivery, predictions_log))
print(f"RMSE (Log Transformation): {rmse_log}")
```

```python
# Square transformation
X_square = np.square(X_delivery)
model_delivery.fit(X_square, y_delivery)
predictions_square = model_delivery.predict(X_square)
rmse_square = np.sqrt(mean_squared_error(y_delivery, predictions_square))
print(f"RMSE (Square Transformation): {rmse_square}")
```

```python
# Square root transformation
X_sqrt = np.sqrt(X_delivery)
model_delivery.fit(X_sqrt, y_delivery)
predictions_sqrt = model_delivery.predict(X_sqrt)
rmse_sqrt = np.sqrt(mean_squared_error(y_delivery, predictions_sqrt))
print(f"RMSE (Square Root Transformation): {rmse_sqrt}")
```

```python
# Visualize the regression line
plt.scatter(df['Sorting Time'], df['Delivery Time'])
plt.plot(df['Sorting Time'], predictions_delivery, color='red')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.title('Sorting Time vs. Delivery Time (Regression Line)')
plt.show()
```
